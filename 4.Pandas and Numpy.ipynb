# Pandas filering and Groupby

```python
pd.read_csv("C:\\Users\\Priyanshu Chauhan\\Downloads\\Salaries.csv")
```

```python
cv = pd.read_csv("C:\\Users\\Priyanshu Chauhan\\Downloads\\Salaries.csv")
```

```python
cv
```

```python
#Extract records where gender 
cv['gender']
```

```python
#Extract records where gender  = Female
cv['gender'] == 'Female'
```

```python
cv[cv['gender'] == 'Female'] # masking
```

```python
cv[cv['gender'] == 'female'] # beacuse f or F
```

```python
#Gender is female and salary is greater then 150000

cv['gender'] == "Female"
cv['salary'] > 150000


cv[(cv['gender'] == "Female") & (cv['salary'] > 150000)]
```

```python
cv[(cv['gender'] == "Female") | (cv['salary'] > 150000)]
```

```python
#What is the average salary of prof?
cv[cv['rank'] == 'Prof']
```

```python
cv[cv['rank'] == 'Prof']['salary']
```

```python
cv[cv['rank'] == 'Prof']['salary'].mean()
```

```python
#Average salary of prof rank and gender is male and female
cv[(cv['rank'] =="Prof")& (cv['gender'] == 'Female')]['salary'].mean()
```

```python
cv[(cv['rank'] =="Prof")& (cv['gender'] == 'Male')]['salary'].mean()
```

```python
cv
```

```python
print(cv[cv['rank'] == 'Prof']['salary'].mean())
print(cv[cv['rank'] == 'AssocProf']['salary'].mean())
print(cv[cv['rank'] == 'AsstProf']['salary'].mean())
```

```python
#groupby method (mean & agg)
```

```python
cv.groupby('rank')['salary'].mean()
```

```python
cv.groupby('rank')['salary'].agg({'mean','median','max', 'min'})
```
